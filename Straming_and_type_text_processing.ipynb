```python
words =["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "finally", "finalized"]
```

```python
print(words)
```

# PorterStremmer

```python
from nltk.stem import PorterStemmer
```

```python
streaming = PorterStemmer()
```

```python
for word in words:
    print(word+ "--->" + streaming.stem(word))
```

```python
streaming.stem("Congratulations")
```

# RegexpStremmer class

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer("ing$|s$|e$|able$", min=4)
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("ingeating")
```

# Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snow_stem = SnowballStemmer("english")
```

```python
for word in words:
    print(word + "--->" + snow_stem.stem(word))
```

```python
streaming.stem("fairly"), streaming.stem("sportingly")
```
