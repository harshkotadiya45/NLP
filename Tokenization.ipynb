```python
corpos = """Hello welcome,to Harsh Kotadiya NLP.
Please do watch entire series! to become expert in NLP
"""
print(corpos)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpos)
```

```python
for sent in documents:
    print(sent)
```

```python
## A paragraph --> Words
## A Sentance --> Words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpos)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpos)
```
