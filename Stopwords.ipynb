```python
paragraph= """Education is one of the most powerful tools for personal and social development. It plays a vital role in shaping an individual’s character, knowledge, and future. Through education, people gain the ability to read, write, communicate effectively, and think logically. It helps individuals understand the world around them and make informed decisions in their daily lives. Education is not limited to books and classrooms; it also teaches values such as discipline, honesty, responsibility, and respect for others.
One of the main benefits of education is that it opens doors to better career opportunities. In today’s competitive world, most professions require specialized knowledge and skills that can only be gained through proper education. A well-educated person has a higher chance of securing a stable job, earning a good income, and achieving financial independence. Education also encourages innovation and creativity, which are essential for economic growth and technological advancement.
Education plays a crucial role in building a strong and progressive society. An educated society is more aware of social issues such as poverty, health, environmental protection, and human rights. Educated citizens are more likely to participate in democratic processes, follow laws, and contribute positively to their communities. They understand the importance of equality, justice, and cooperation, which helps in maintaining peace and harmony in society.
Moreover, education improves the overall quality of life. Educated people are more conscious about their health and hygiene, leading to healthier lifestyles. They are better equipped to understand medical information, follow safety guidelines, and adopt preventive measures against diseases. Education also empowers individuals, especially women, by giving them confidence and independence. When women are educated, they can make better decisions for themselves and their families, which leads to the development of the entire nation."""
```

```python
paragraph
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download("stopwords")
```

```python
stopwords.words("english")
```

```python
stemmer = PorterStemmer()
```

```python
sentences = nltk.sent_tokenize(paragraph)
type(sentences)
```

```python
## Apply Stopwords and filter and then Apply Stemming
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word.lower() not in set(stopwords.words("english"))]
    sentences[i] = " ".join(words) #converting all List of words the words in sentences.
```

```python
sentences
```

```python
from nltk.stem import SnowballStemmer
```

```python
snow_ball = SnowballStemmer("english")
```

```python
## Apply Stopwords and filter and then Apply Snowball Stemming
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [snow_ball.stem(word) for word in words if word.lower() not in set(stopwords.words("english"))]
    sentences[i] = " ".join(words) #converting all List of words the words in sentences.
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
## Apply Stopwords and filter and then Apply Stemming
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word, pos='v') for word in words if word.lower() not in set(stopwords.words("english"))]
    sentences[i] = " ".join(words) #converting all List of words the words in sentences.
```
