```python
sentence = "The Eiffel Tower Was Build from 1887 to 1889 by Freanch enginner Gustave Eiffel, Whose Company specified in building metal frameworks and structures."
```

```python
import nltk
words = nltk.word_tokenize(sentence)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
tag_eliments = nltk.pos_tag(words)
tag_eliments
```

```python
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```
